The DifferentialEquations.jl algorithms choose the number type of their calculation given their input. Use this fact to solve the Lorenz equation using BigFloats. Make a 3D plot of the Lorenz attractor using the plot recipe.

```julia
using DifferentialEquations, Plots
gr()
```

```julia
function lorenz!(du, u, p, t)
    σ, ρ, β = p
    du[1] = σ * (u[2] - u[1])
    du[2] = u[1] * (ρ - u[3]) - u[2]
    du[3] = u[1] * u[2] - β * u[3]
end
```

```julia
u0 = [BigFloat(1.0), BigFloat(0.0), BigFloat(0.0)]
tspan = (0.0, 100.0)
p = (10, 28, 8 / 3)
```

```julia
prob = ODEProblem(lorenz!, u0, tspan, p)
sol = solve(prob);
```

```julia
plot(sol)
```

```julia
plot(sol, vars=(1, 2, 3))
```
